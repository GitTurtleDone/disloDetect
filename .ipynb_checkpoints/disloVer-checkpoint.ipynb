# Processing

```python
def readBB(filename, dataOrg = 'row'): # returns coordinates of bounding boxes stored in a .txt file 
  # if dataOrg = 'row' each row will have class and coordinates of a bounding box
  # if dataOrg = 'col' each columns will store class and coordinates of all corresponding bounding boxes
  try:
    with open(filename, 'r') as file:
      data = []
      for line in file:
        # Split the line into a list of numbers (as strings)
        numbers_as_str = line.strip().split()
        # Check if there are exactly five numbers
        if len(numbers_as_str) != 5:
          raise ValueError(f"Expected 5 numbers per row in {filename}, found {len(numbers_as_str)} in line: {line}")
        # Convert strings to floating-point numbers
        data.append([float(num) for num in numbers_as_str])
    if dataOrg == 'row':
        return data
    else: 
        # Organize data into columns using list comprehension
        columns = [[row[i] for row in data] for i in range(5)]
        return columns

  except FileNotFoundError:
    print(f"Error: File '{filename}' not found.")
    return None
  except ValueError as e:
    print(e)
    return None
def getTxtFiles(folderPath):
    filenames = []
    for filename in os.listdir(folderPath):
        print(filename)
        if filename.endswith(".txt"):
            filenames.append(filename)
    return filenames
def drawBB(image, coordinates, convertFromXYWHN = True):
    """
    Draws a bounding box on an image given coordinates.

    Args:
      image (numpy.ndarray): The image as a NumPy array.
      coordinates (list): A list containing four elements representing
                          top-left x, top-left y, bottom-right x, and bottom-right y
                          coordinates of the bounding box.
    """
    if convertFromXYWHN:
        xTL = coordinates[0] - coordinates[2]/2
        xBR = coordinates[0] + coordinates[2]/2
        yTL = coordinates[1] - coordinates[3]/2
        yBR = coordinates[1] + coordinates[3]/2
    else:
        (xTL, yTL, xBR, yBR) = coordinates
    xTL = int(xTL * image.size[0])
    xBR = int(xBR * image.size[0])
    yTL = int(yTL * image.size[1])
    yBR = int(yBR * image.size[1])
    bbox = (xTL, yTL, xBR, yBR)
    draw = ImageDraw.Draw(image)
    draw.rectangle(bbox, outline = 'red', width = 2)  # Green box with thickness 2
def sumBBHeights(labelFilePath):
    dataBB = readBB(labelFilePath, dataOrg = 'col')
    #print(dataBB[4])
    return (sum (dataBB[4]))
def getAllSums(filePaths):
    #fileNames = getTxtFiles(labelFolderPath)
    return ([sumBBHeights(filePath) for filePath in filePaths])
```

```python
labelFolderPath = './ProcessingData/predict66/labels/'
fileNames = getTxtFiles(labelFolderPath)
filePaths = [(labelFolderPath + fileName) for fileName in fileNames]
preAllSums = getAllSums(filePaths)
print(preAllSums)
```

```python
arrImageNum = range(1,21)
testDislo2AllSums = np.array(testDislo2AllSums)
preAllSums = np.array(preAllSums)
err = (preAllSums-testDislo2AllSums)/testDislo2AllSums*100
csvFilePath = labelFolderPath + "pre66AllSums.csv" 
with open(csvFilePath, "w", newline="") as csvfile:
  # Create a csv writer object
  writer = csv.writer(csvfile)
  # Combine data using zip and write as rows
  writer.writerow(["File name","Image number", "Human-labeled", "Predicted", "Error %"])
  writer.writerows(zip(fileNames, arrImageNum, testDislo2AllSums, preAllSums, err))
fig, ax = plt.subplots(1,2)
line1, = ax[0].plot(arrImageNum, testDislo2AllSums, marker = 's', markersize=10, markerfacecolor='none', markeredgewidth=1, markeredgecolor='black', linestyle = 'None', label = 'Human labeled 2nd DS')
line2, = ax[0].plot(arrImageNum, preAllSums, marker = 'o', markersize=10, markerfacecolor='none', markeredgewidth=1, markeredgecolor='red', linestyle = 'None', label = 'predicted by 2nd model')
line3, = ax[1].plot(arrImageNum, err, marker = '^', markersize=10, markerfacecolor='none', markeredgewidth=1, markeredgecolor='green', linestyle = 'None', label = 'predicted by 2nd model')
```
